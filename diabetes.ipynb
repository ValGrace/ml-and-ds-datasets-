```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/ValGrace/ml-and-ds-datasets-/main/diabetes.csv'
diabetes = pd.read_csv(url)
diabetes.head()
```

```python
X = diabetes.iloc[:, 0:8].values
Y = diabetes.iloc[:, -1].values
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = 100)
```

```python
from sklearn.tree import DecisionTreeClassifier 
tree_model = DecisionTreeClassifier(criterion='entropy', max_depth=5) 
tree_model.fit(X_train, y_train)
print('training set accuracy: {:.3f}' .format(tree_model.score(X_train, y_train)))
print('test set accuracy: {:.3f} ' .format(tree_model.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
diabetes_mod = export_graphviz(tree_model, out_file='tree_model.dot',
                class_names=['Diabetic', 'Non-diabetic'],
                feature_names=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
from sklearn import tree
diabetes_diagram = plt.figure(figsize=(25, 15))
tree.plot_tree(tree_model, filled=True,
                class_names=['Diabetic', 'Non-diabetic'],
                feature_names=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
```
