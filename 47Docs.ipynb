```python
from PIL import Image
from paddleocr import PaddleOCR
```

```python
ocr = PaddleOCR(use_textline_orientation=True, lang='en')
```

```python
import io
def extract_text_from_pdf(input_file):
  doc = fitz.open(input_file)
  all_text = []

  for page in doc:
    text = page.get_text()
    if text.strip():
      all_text.append(text)
    else:
      pix = page.get_pixmap(dpi=300)
      img = Image.open(io.BytesIO(pix.tobytes("png")))
      result = ocr.ocr(img, cls=True)
      page_text = [line[1][0] for line in result[0]]
      all_text.append('\n'.join(page_text))
  return '\n\n'.join(all_text)
```

```python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get("GEMINI_API_KEY"))
```

```python
def translate_text(text, target_language="swahili"):
  prompt = f"Translate the following text to {target_language}:\n\n{text}"
  response = client.models.generate_content(
      model="gemini-2.5-flash",

      contents = prompt
  )
  return response.text
```

```python
from fpdf import FPDF
import os

def build_pdf(translated_text, output_path):
  pdf = FPDF()
  pdf.add_page()

  font_name = 'DejaVuSans'
  local_font_path = '/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf'

  font_exists = os.path.exists(local_font_path)
  print(f"Font exists at {local_font_path}: {font_exists}")

  try:
    if font_exists:
      # Removed uni=True as it's deprecated; fpdf2 usually handles unicode automatically with proper fonts
      pdf.add_font(font_name, '', local_font_path)
      pdf.set_font(font_name, '', 10) # Reduced font size from 12 to 10
    else:
      raise RuntimeError(f"DejaVuSans font not found at {local_font_path}.")
  except RuntimeError as e:
    print(f"Error adding font: {e}. Falling back to default font, which may not support all characters.")
    pdf.set_font("Arial", size=10) # Fallback to Arial if font addition fails, also reduced to 10

  # Calculate available width: Page width - left margin - right margin
  # Default A4 width is 210mm, default margins are 10mm each.
  page_width = pdf.w
  left_margin = pdf.l_margin
  right_margin = pdf.r_margin
  # Provide a small buffer, e.g., 10mm less than full available width
  cell_width = page_width - left_margin - right_margin - 10

  for line in translated_text.split('\n'):
    # Ensure multi_cell can wrap if lines are very long
    # Use the calculated cell_width instead of 0, and adjusted line height
    pdf.multi_cell(cell_width, 8, line)
  pdf.output(name=output_path + ".pdf", dest='F')
  files.download(output_path + ".pdf")
```

```python
def replace_text(pdf_path, output_path, target_language):
  doc = fitz.open(pdf_path)

  for page_index, page in enumerate(doc):
    blocks = page.get_text("blocks")
    for b in blocks:
      rect = fitz.Rect(b[0], b[1], b[2], b[3])
      translated = translate_text(b[4], target_language)
      page.insert_textbox(rect, translated, fontsize=12, fontname="helv", align=0)
  doc.save(output_path + ".pdf")
  # files.download(output_path + ".pdf"
```
